```python
Dataset=pd.read_csv("census-income.data.csv")
Dataset.info()
WC=pd.get_dummies(Dataset["workclass"],drop_first=True)
Dataset=pd.concat([WC,Dataset],axis=1)
Dataset.drop(["workclass"],axis=1,inplace=True)
```

```python
ED=pd.get_dummies(Dataset["education"],drop_first=True)
Dataset=pd.concat([Dataset,ED],axis=1)
Dataset.drop(["education"],axis=1,inplace=True)
```

```python
MS=pd.get_dummies(Dataset["marital_status"],drop_first=True)
Dataset=pd.concat([Dataset,MS],axis=1)
Dataset.drop(["marital_status"],axis=1,inplace=True)
```

```python
OCC=pd.get_dummies(Dataset["occupation"],drop_first=True)
Dataset=pd.concat([Dataset,OCC],axis=1)
Dataset.drop(["occupation"],axis=1,inplace=True)
```

```python
RS=pd.get_dummies(Dataset["relationship"],drop_first=True)
Dataset=pd.concat([Dataset,RS],axis=1)
Dataset.drop(["relationship"],axis=1,inplace=True)
```

```python
Race=pd.get_dummies(Dataset["race"],drop_first=True)
Dataset=pd.concat([Dataset,Race],axis=1)
Dataset.drop(["race"],axis=1,inplace=True)
```

```python
SEX=pd.get_dummies(Dataset["sex"],drop_first=True)
Dataset=pd.concat([Dataset,SEX],axis=1)
Dataset.drop(["sex"],axis=1,inplace=True)
```

```python
NC=pd.get_dummies(Dataset["native_country"],drop_first=True)
Dataset=pd.concat([Dataset,NC],axis=1)
Dataset.drop(["native_country"],axis=1,inplace=True)
```

```python
X=Dataset.drop('salary',axis=1)
y=Dataset['salary']
```

```python
sal=pd.get_dummies(Dataset["salary"])
y=pd.concat([y,sal],axis=1)
y.drop(["salary"],axis=1,inplace=True)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X.iloc[:,0:100])
X.iloc[:,0:100]=imputer.transform(X.iloc[:,0:100])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train.iloc[:,0:100] = sc.fit_transform(X_train.iloc[:,0:100])
X_test.iloc[:,0:100] = sc.transform(X_test.iloc[:,0:100])
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy',random_state = 0)
classifier.fit(X_train,y_train)
cls_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc1=accuracy_score(y_test,cls_pred)
print(acc1)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier1 = RandomForestClassifier(n_estimators=10,criterion='entropy',random_state = 0)
classifier1.fit(X_train,y_train)
cls_pred1=classifier1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc2=accuracy_score(y_test,cls_pred1)
print(acc2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier3 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier3.fit(X, y)
cls_pred3=classifier3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc3=accuracy_score(y_test,cls_pred3)
print(acc3)
```
